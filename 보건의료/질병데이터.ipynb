```python
import requests
import pandas as pd
```

```python
dflist = []
dissCd = ['1', '2', '3', '4', '5', '15']
znCd = ['11', '26', '27', '28', '29', '30', '31', '41', '42', '43', '44', '45', '46', '47', '48', '49']
url = 'http://apis.data.go.kr/B550928/dissForecastInfoSvc/getDissForecastInfo'

for i in dissCd:
    for j in znCd:
        params = {
            'serviceKey': 'oyxJfzSFwRSWbZS4ceoPwqKklZRfDFFP+5EYvDLIKMRQwhSDy6DDS0HnsU7fSzcyEPSkr0GbJIx0RUmN/4kUbg==',  # 올바른 서비스 키를 입력하세요.
            'numOfRows': '75',
            'pageNo': '1',
            'type': 'json',
            'dissCd': i,
            'znCd': j
        }

        response = requests.get(url, params=params)
        response_content = response.content.decode('utf-8')  # 응답 내용을 디코딩

        # 응답 데이터가 JSON 형식인지 확인
        if response.status_code == 200 and response_content:
            try:
                data = response.json()
                for k in range(len(data['response']['body']['items'])):
                    dflist.append(data['response']['body']['items'][k])
            except ValueError:
                print(f"Failed to decode JSON for dissCd={i}, znCd={j}: {response_content}")
        else:
            print(f"Failed to retrieve data for dissCd={i}, znCd={j}: {response.status_code}")
```

# Mysql에 저장

```python
from sqlalchemy import create_engine

# CSV 파일 경로
csv_file_path = 'disease_predict_data_' + todate + '.csv'

# MySQL 연결 정보
user = 'root'
password = 'mysql123'
host = '127.0.0.1:3306'
database = 'DSDT'
table_name = 'disease_predict_data'

# SQLAlchemy 엔진 생성
engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}/{database}')
```

```python
# CSV 파일 읽기
df = pd.read_csv(csv_file_path, encoding='utf-8')

# 데이터프레임을 MySQL 테이블로 저장
df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

print(f"Data successfully saved to {database}.{table_name}")
```
